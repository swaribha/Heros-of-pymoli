```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#Counting the unique number of players
total_players=purchase_data['SN'].nunique()
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Calculating the unique number of items 
unique_item_count=purchase_data['Item ID'].nunique()
print(unique_item_count)
#Average price per item
Average_price=round(purchase_data['Price'].mean(),2)
print(Average_price)
#Total number of purchases
total_purchase=purchase_data['Purchase ID'].count()
print(total_purchase)
#Total revenue Generated
total_revenue=purchase_data['Price'].sum()
print(total_revenue)
```

```python
#
dict_summary={"Number of Unique Items":unique_item_count,
              "Average Price": Average_price,
              "Number of Purchases":total_purchase,
              "Total Revenue":total_revenue}
purchase_summary=pd.DataFrame([dict_summary])
purchase_summary['Total Revenue'].map
purchase_summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#For Gender Analysis we just need unique player
#Creating a dataframe with unique screen name to remove duplicates in Gender
gender_df=purchase_data.drop_duplicates('SN')
#Counting the Number of Male female and Non-disclosed Players
gender_analysis_df=pd.DataFrame(gender_df['Gender'].value_counts())
#Adding percentage of gender to the Analysis
gender_analysis_df['Percentage']=round((gender_analysis_df['Gender']/gender_analysis_df['Gender'].sum()*100),2)
gender_analysis_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Getting group by series by Gender 
purchase_analysis=purchase_data.groupby('Gender')
#.agg({'SN':{'Purchase Count':"count"},'Price':{'Average Purchase Price':"mean",'Total Purchase Value':"sum"}})

#  * Purchase Count
Purchase_Count=purchase_analysis['SN'].count()
purchase_analysis_summary=pd.DataFrame(Purchase_Count)
#* Average Purchase Price
purchase_analysis_summary['Average Purchase Price']=purchase_analysis['Price'].mean()
# * Total Purchase Value
purchase_analysis_summary['Total Purchase Value']=purchase_analysis['Price'].sum()
#Count the unique gender for Avg purchase per person
unique_gender_count=pd.Series(purchase_data.groupby('Gender')['SN'].nunique())
# print(unique_gender_count)
purchase_analysis_summary['Avg Total Purchase per Person']=purchase_analysis_summary['Total Purchase Value']/unique_gender_count
#Rename column SN to Purchase count
purchase_analysis_summary.rename(columns={'SN':'Purchase Count'},inplace=True)
#Formating the data
purchase_analysis_summary[['Average Purchase Price','Total Purchase Value','Avg Total Purchase per Person']]=purchase_analysis_summary[['Average Purchase Price','Total Purchase Value','Avg Total Purchase per Person']].applymap("${:,.2f}".format)
purchase_analysis_summary
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#checking age range for creating the bins
age_min=purchase_data['Age'].min()
age_max=purchase_data['Age'].max()
#creating the bins for defining the age group 
bins=[0,10,15,20,25,30,35,40,45]
#labels for bins
labels=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
#creating dataframe for age analysis by dropping duplicates
age_anaylsis=purchase_data.drop_duplicates('SN')
# adding the "Age group " 
age_anaylsis["Age Group"]=pd.cut(age_anaylsis['Age'],bins,labels=labels,right=False)
# Calulating the total Count by Age group
purchase_count_age=age_anaylsis["Age Group"].value_counts()
# Creating the summary Dataframe
age_anaylsis_summary=pd.DataFrame(purchase_count_age)
age_anaylsis_summary.rename(columns={'Age Group':'Total Count'},inplace=True)
#Calculating the percentage of player by age group
age_anaylsis_summary['Percentage of Players']=age_anaylsis_summary['Total Count']/age_anaylsis_summary['Total Count'].sum()*100
age_anaylsis_summary['Percentage of Players']=age_anaylsis_summary['Percentage of Players'].map('{:.2f}%'.format)
#Sorting the dataframe based on index
age_anaylsis_summary.sort_index(inplace=True)
age_anaylsis_summary
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#copying 
purchase_analysis_age=purchase_data.copy()
purchase_analysis_age['Age Group']=pd.cut(purchase_analysis_age['Age'],bins,labels=labels,right=False)
# print(purchase_analysis_age.head())
# Calulating the purchase Count by Age group
purchase_groupby=purchase_analysis_age.groupby("Age Group")
purchase_count_age=purchase_groupby["Age Group"].count()
# #Creating the summary Dataframe
purchase_age_anaylsis_summary=pd.DataFrame(purchase_count_age)
purchase_age_anaylsis_summary.rename(columns={'Age Group':'Purchase Count'},inplace=True)
# #Calculating the percentage of player by age group
purchase_age_anaylsis_summary['Average Purchase Price']=purchase_groupby['Price'].mean()
purchase_age_anaylsis_summary['Total Purchase Value']=purchase_groupby['Price'].sum()
#calculating unique purchase count in each group
unique_age_group=pd.Series(purchase_groupby['SN'].nunique())
#print(unique_age_group)
purchase_age_anaylsis_summary['Avg Total Purchase per Person']=purchase_age_anaylsis_summary['Total Purchase Value']/unique_age_group
#Formating the data
purchase_age_anaylsis_summary[['Average Purchase Price','Total Purchase Value','Avg Total Purchase per Person']]=purchase_age_anaylsis_summary[['Average Purchase Price','Total Purchase Value','Avg Total Purchase per Person']].applymap("${:,.2f}".format)
purchase_age_anaylsis_summary
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#making a copy of original data for analysis
top_spenders_analysis=purchase_data.copy()
#creating a group by object for name
top_spenders_groupby=top_spenders_analysis.groupby('SN')
# Getting count of purchases by each user
top_spenders_count=top_spenders_groupby['Purchase ID'].count()
#Creating a summary data frame
top_spenders_analysis_summary=pd.DataFrame(top_spenders_count)
#Renaming the Purchase Id to Purchase Count
top_spenders_analysis_summary.rename(columns={'Purchase ID':'Purchase Count'},inplace=True)
#Average Purchase Price by user
top_spenders_analysis_summary['Average Purchase Price']=top_spenders_groupby['Price'].sum()/top_spenders_analysis_summary['Purchase Count']
top_spenders_analysis_summary['Total Purchase Value']=top_spenders_groupby['Price'].sum()
# top_spenders_analysis_summary
top_five_spenders=top_spenders_analysis_summary.nlargest(5,'Total Purchase Value')
#Applying formatting
top_five_spenders[['Average Purchase Price',
                   'Total Purchase Value']]=top_five_spenders[['Average Purchase Price',
                                                               'Total Purchase Value']].applymap("${:,.2f}".format)
# top_spenders_analysis.head()
top_five_spenders
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# creating new dataframe for Item analysis
item_df=purchase_data[['Item ID','Item Name','Price']]
# dropping duplicate names and item number
item_df.drop_duplicates(inplace=True)
#Creating a group by object for item analysisPurchase ID	SN	Age	Gender	Item ID	Item Name	Price
item_df_groupby=purchase_data.groupby('Item ID')
# counting the Purchase count by item id
Purchase_count_item=item_df_groupby['Purchase ID'].count()
# Purchase_count_item
item_analysis=pd.DataFrame(Purchase_count_item)
item_analysis.rename(columns={'Purchase ID':'Purchase Count'},inplace=True)
# merging the dataframe with item list
merge_summary=pd.merge(item_df,item_analysis,on='Item ID',how='inner')
merge_summary.set_index('Item ID',inplace=True)
# Calculating the total Purchase for item
merge_summary['Total Purchase Value']=merge_summary['Purchase Count']*merge_summary['Price']
top_five_items=merge_summary.nlargest(5,'Purchase Count')
#Formating the Data
top_five_items[['Price','Total Purchase Value']]=top_five_items[['Price',
                                                                 'Total Purchase Value']].applymap("${:,.2f}".format)
top_five_items
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# extracting top five profitable items
most_profitable_items=merge_summary.nlargest(5,'Total Purchase Value')
# Formatting the data
most_profitable_items[['Price','Total Purchase Value']]=most_profitable_items[['Price',
                                                                 'Total Purchase Value']].applymap("${:,.2f}".format)
most_profitable_items
```
